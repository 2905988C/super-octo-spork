```python
country=df['country']
country.reset_index(drop=True,inplace=True)
print(country)
```

```python
se=pd.Series(country)
count=se.value_counts()
count=list(count)
print(count)
```

```python
country_name=dict(se.value_counts())
country_name=list(country_name)
print(country_name)
```

```python
plt.bar(country_name,count)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x=country_name, y=count)
```

```python
# Q2
```

```python
plt.figure(1, figsize=(8,6))
plt.suptitle('Counts of the most visited countries', fontsize=14)
plt.xlabel('country')
plt.ylabel('count')
plt.xticks(size =6)
plt.xticks(rotation=45)
plt.ylim(0,700)
ax = plt.bar(country_name,count, color='orange', alpha=.5, edgecolor='maroon')
plt.show()
```
